```python
import matplotlib.pyplot as plt
import numpy as np

# Set the font to Times New Roman
plt.rc('font', family='Times New Roman', style='italic')

# Intersection point
x = 1
y = 2

# Plot the lines
fig, ax = plt.subplots()

ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')

# Set axis limits
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

# Add grid lines and set their color
ax.grid(True, color='lightgrey')

# mark the origin
ax.plot(0, 0, 'ko', markersize=5)
ax.text(-0.5, -0.5, 'O', fontsize=12)

# Add axis arrows
ax.arrow(0, -4, 0, 8, length_includes_head=True, head_width=0.2)
ax.arrow(-4, 0, 8, 0, length_includes_head=True, head_width=0.2)

# Add x, y labels to the end of arrows
ax.text(3.9, -0.5, 'x', fontsize=12)
ax.text(-0.5, 3.9, 'y', fontsize=12)

# Set the outer box border to grey
for spine in ax.spines.values():
    spine.set_edgecolor('lightgrey')
# Modify tick label colors for x and y axes
for tick in ax.xaxis.get_ticklabels():
    tick.set_color('grey')
for tick in ax.yaxis.get_ticklabels():
    tick.set_color('grey')

# Draw the vectors
ax.arrow(0, 0, 2, -1, color="blue", head_width=0.2, head_length=0.2)
ax.arrow(0, 0, -1, 2, color="red", head_width=0.2, head_length=0.2)
ax.arrow(2, -1, -2, 4, color="red", head_width=0.2, head_length=0.2, linestyle='dashed')
ax.arrow(0, 0, 0, 3, color="green", head_width=0.2, head_length=0.2, linewidth=4, alpha=0.5)

# Annotate the vectors
ax.annotate('vector 1', xy=(1.5, -0.5), color='blue', fontsize=12)
ax.annotate('vector 2', xy=(-2, 1.5), color='red', fontsize=12)
ax.annotate('2*vector 2', xy=(0.5, 3), color='red', fontsize=12)
ax.annotate('target vector', xy=(-1.7, 3), color='green', fontsize=12)

plt.savefig('L1_2.jpg', dpi=400)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
def plane1(x, y):
    return 2*x - y

def plane2(x, y):
    return (-x + 2*y + 1) / 1.0

def plane3(y, z):
    return (4 + 3*y) / 4.0

# 
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z1 = plane1(X, Y)
Z2 = plane2(X, Y)
Z3 = plane3(Y, X)  # Xz

# 
from scipy.linalg import solve
A = [[2, -1, 0], [-1, 2, -1], [0, -3, 4]]
b = [0, -1, 4]
intersection = solve(A, b)

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
ax.plot_surface(X, Y, Z1, alpha=0.5, facecolors='red', rstride=100, cstride=100)
ax.plot_surface(X, Y, Z2, alpha=0.5, facecolors='g', rstride=100, cstride=100)
ax.plot_surface(Y, Z3, X, alpha=0.5, facecolors='y', rstride=100, cstride=100)

# 
ax.scatter(*intersection, color="r", s=100, label=f'Intersection {tuple(intersection)}')

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 
ax.text(0, 0, 10, r"$2x - y = 0$", color='red')
ax.text(0, 10, 0, r"$-x + 2y - z = -1$", color='green')
ax.text(10, 0, 0, r"$-3y + 4z = 4$", color='blue')

# 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
v1 = np.array([2, -1, 0])
v2 = np.array([-1, 2, -3])
v3 = np.array([0, -1, 4])

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
# x, y, z 01
# x, y, z
x, y, z = 1, 1, 1
result = x*v1 + y*v2 + z*v3

# 
ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='b', label='v1 (2, -1, 0)')
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='r', label='v2 (-1, 2, -3)')
ax.quiver(0, 0, 0, v3[0], v3[1], v3[2], color='g', label='v3 (0, -1, 4)')
ax.quiver(0, 0, 0, result[0], result[1], result[2], color='k', label='Linear Combination', linestyle='dashed')

# 
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])
ax.set_zlim([-2, 2])

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 
ax.legend()

# 
plt.savefig('L1_3.jpg', dpi=400)
plt.show()
```
